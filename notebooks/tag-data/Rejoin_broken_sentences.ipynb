```python
import matplotlib.pyplot as plt 
import numpy as np
import os 
import random
import sys
import pandas as pd
import csv
import re
import time
```

```python
#read in CSV for  for full corpus
corpus = pd.read_csv('data/interim/extended_podcast_sentences.csv')

print(f'Loaded {corpus.shape[0]} sentences')
```

```python
# Loop through all rows to find sentences that were split.
i = 0
first_part = corpus.string[0]
first_source = corpus.source[0]

rejoined = 0
outputDf = pd.DataFrame()
while i < corpus.shape[0]-1:
    first_part = corpus.string[i]
    first_source = corpus.source[i]
    conf = corpus.confidence[i]
    term = first_part[-1]
    n = 1
    while (term not in ['.','!','?']) & (first_source == corpus.source[i+1]):
            second_part = corpus.string[i+1]
            conf += corpus.confidence[i+1]
            first_part = first_part + " " + second_part
            #print(first_part,i)
            rejoined += 1
            n += 1
            term=first_part[-1]
            i += 1 # skip over second part in loop
    conf = conf/n
    outputDf = outputDf.append( {"string":first_part, "confidence":conf, "source":corpus.source[i]},
                    ignore_index=True)
    i = i + 1   
    if i % 10000 == 0:
        print(f'{time.asctime()} line {i} rejoined {rejoined}')
        outputDf.to_csv('data/interim/rejoined_extended_podcast_sentences.csv', sep=',',index=False)

#print(outputDf)
print(f'Rejoined {rejoined} strings'  )    
```

```python
# output dataframe to file
outputDf.to_csv('data/interim/rejoined_extended_podcast_sentences.csv', sep=',',index=False)
```
